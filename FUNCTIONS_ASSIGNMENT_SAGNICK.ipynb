5) What are iterators in Python and how do they differ from iterables?

Ans - Iterables are objects you can loop over (like lists, strings, tuples), while iterators are objects that keep track of where you are in a loop and return items one at a time using the next() function.

Difference:

6) Explain the concept of generators in Python and how they are defined.

Ans - Generators in Python are a way to create iterators using a function that yields values one at a time, instead of returning all values at once.

Generators are defined like regular functions but use the yield keyword instead of return.

```python
#Example

def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

gen = count_up_to(3)
print(next(gen))  # 1
print(next(gen))  # 2
print(next(gen))  # 3
```

7) What are the advantages of using generators over regular functions?

Ans - Advantages of using generators over regular functions:

Memory Efficient – They yield one value at a time, so they don’t store the whole result in memory.

Faster for Large Data – They start producing values immediately, no need to wait for all processing.

Lazy Evaluation – Values are only computed when needed.

Simpler Code for Iteration – Generators make writing custom iterators easier and cleaner.

Infinite Sequences – Useful for representing endless data streams (like sensor data or logs).

```python
#Example -

def count():
    n = 0
    while True:
        yield n
        n += 1

gen = count()
print(next(gen))  # 0
print(next(gen))  # 1
```

10) What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Ans - The functions map(), reduce(), and filter() in Python are all used for functional-style programming. Each one serves a different purpose in processing data from iterables.

1. map(function, iterable)

Purpose: Transforms each element in an iterable by applying a function.

Returns: A new iterable with transformed values.

```python
#Example -

numbers = [1, 2, 3, 4]
result = map(lambda x: x * 2, numbers)
print(list(result))
```

2. filter(function, iterable)
Purpose: Filters elements from an iterable for which the function returns True.

Returns: A new iterable with only the filtered values.

```python
#Example -

numbers = [1, 2, 3, 4]
result = filter(lambda x: x % 2 == 0, numbers)
print(list(result))
```

3.  reduce(function, iterable)

Purpose: Reduces the iterable to a single cumulative value by applying the function cumulatively.

Returns: A single value.



```python
from functools import reduce

numbers = [1, 2, 3, 4]
result = reduce(lambda x, y: x + y, numbers)
print(result)
```

5) Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.a
fib = FibonacciIterator(10)
for num in fib:
    print(num, end=' ')

```

7)  Implement a generator function that reads a file line by line and yields each line as a string.

```python
# Define the generator function
def read_file_lines(file_path):
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.rstrip('\n')  # Yield each line without newline
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")

# Example usage
if __name__ == "__main__":
    for line in read_file_lines('example.txt'):
        print(line)
```
